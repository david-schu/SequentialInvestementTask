```python
def chi_square(c_data, index):
    
    chi_squares = {}
    
    # portfolio delta
    chi_squares['d_portfolio'] = chi2_contingency(pd.crosstab(np.sign(c_data['d_portfolio']), np.sign(c_data['d_bet'])))[0]
    
    # market delta
    chi_squares['d_market'] = chi2_contingency(pd.crosstab(np.sign(c_data['d_market']), np.sign(c_data['d_bet'])))[0]
    
    # market delta
    chi_squares['d_market_belief'] = chi2_contingency(pd.crosstab(np.sign(c_data['d_market_belief']), np.sign(c_data['d_bet'])))[0]
    

    chi_squares = pd.DataFrame(chi_squares, index=[index])
    
#     if index == 0:
#         dfi.export(pd.crosstab(np.sign(c_data['d_portfolio']), np.sign(c_data['d_bet'])), 'crosstan_d_portfolio.png')
#         dfi.export(pd.crosstab(np.sign(c_data['d_market']), np.sign(c_data['d_bet'])), 'crosstan_d_market.png')
#         dfi.export(pd.crosstab(np.sign(c_data['d_market_belief']), np.sign(c_data['d_bet'])), 'crosstan_d_market_belief.png')        
#         dfi.export(pd.crosstab(bet_size, np.sign(c_data['d_bet'])), 'crosstab_counterfactual_regret.png')  
#         dfi.export(chi_squares.reindex(index=['chi_square']), 'chi_square.png')
    
    return chi_squares

def pearsonr_pval(x,y):
        return pearsonr(x,y)[1]
```

# Clustering subjects betsizes


```python
np.random.seed(123)

bet_data = pd.crosstab(all_data['subj'], all_data['bet'])

lowest_bic = np.infty
bic = []
n_components_range = range(1, 8)

for n_components in n_components_range:
    # Fit a Gaussian mixture with EM
    gmm = mixture.GaussianMixture(
        n_init=1000,
        n_components=n_components, 
        reg_covar=1,
        init_params='random'
    )
    gmm.fit(bet_data)
    bic.append(gmm.bic(bet_data))
    if bic[-1] < lowest_bic and bic[-1]>0:
        lowest_bic = bic[-1]
        best_gmm = gmm

preds = best_gmm.predict(bet_data)
len(best_gmm.means_)
```

```python
for subj in np.unique(all_data['subj']):
    all_data.loc[all_data['subj']==subj, 'cluster'] = int(preds[subj])
```

# Linear Regression

```python
regression_results= pd.DataFrame()
p_values = pd.DataFrame()

data = load_data('all')
for i, subject in enumerate(tqdm(data)):

    for column in subject.columns:
        if abs(subject[column]).sum()>0:
            subject[column] = zscore(subject[column], nan_policy='omit')
    
    md = sm.OLS.from_formula('d_bet ~ 0  + d_market_belief + d_market_uncertainty + regret + rejoice + counterfactual_rejoice + counterfactual_regret', subject, missing='drop')
    result = md.fit()
    
    res = result.params
    res['ll'] = result.llf
    res['bic'] = result.bic
    
    regression_results = regression_results.append(res, ignore_index=True)
    p_values= p_values.append(result.pvalues, ignore_index=True)
```

```python
fig, ax = plt.subplots(1,1, figsize=(figwidth,3), dpi=300, sharey=True, sharex=True)

plt.figure(dpi=400)
props = {'color': 'tab:blue'}
plot_params = regression_results[['d_market_belief', 'd_market_uncertainty', 'rejoice', 'regret', 'counterfactual_rejoice', 'counterfactual_regret']]
plot_params.boxplot(notch=True,positions=np.arange(len(plot_params.columns))*2-0.3, widths=.4, medianprops=props, capprops=props, whis=[5,95], showfliers=False, ax=ax)

props = {'color': 'tab:orange'}
plot_p = p_values[['d_market_belief', 'd_market_uncertainty', 'rejoice', 'regret', 'counterfactual_rejoice', 'counterfactual_regret']]
plot_p.boxplot(notch=True,positions=np.arange(len(plot_params.columns))*2+0.3, widths=.4, boxprops=props, medianprops=props, whiskerprops=props, capprops=props, whis=[5,95], showfliers=False, ax = ax)

ax.grid(False)
ax.hlines(0, -0.5, len(plot_params.columns)*2, 'k', linestyles='dashed')
ax.set_xticks(np.arange(len(plot_params.columns))*2,)
ax.set_xticklabels(plot_params.columns, rotation=45, ha="right")
ax.legend(handles=[mpatches.Patch(color='tab:blue', label='coefficient'), mpatches.Patch(color='tab:orange', label='p_value')])
ax.hlines(0.05, -0.5, len(plot_params.columns)*2, 'tab:orange')


plt.show()
```

```python
data = load_data('all')
all_data = pd.DataFrame()


for i, subject in enumerate(data):
    subject['subj'] = i
    subject['cluster'] = preds[i]
    
    all_data = all_data.append(subject)
all_data.index = np.arange(len(all_data))
for column in all_data.columns:
    if column in ['subj', 'cluster'] or abs(all_data[column]).sum()==0:
        continue
    all_data[column] = zscore(all_data[column], nan_policy='omit')

all_data0 = all_data[all_data['cluster']==0]
all_data1 = all_data[all_data['cluster']==1]
```

```python
bets = np.arange(0,110,10)
p_inc = np.zeros(len(bets))
p_dec = np.zeros(len(bets))
p_stay = np.zeros(len(bets))

for i, bet in enumerate(bets):
    rel_bets = all_data[all_data.bet==bet]
    p_inc[i] = (rel_bets.d_bet>0).sum() / len(rel_bets)
    p_dec[i] = (rel_bets.d_bet<0).sum() / len(rel_bets)
    p_stay[i] = (rel_bets.d_bet==0).sum() / len(rel_bets)

plt.plot(bets, p_inc, label='p(increase bet)')
plt.plot(bets, p_dec, label='p(decrease bet)')
# plt.plot(bets, p_stay, label='p(stay with bet)')
plt.xlabel('betsize')
plt.ylabel('p')
plt.legend()
```

# Check for state dependency

```python
data = load_data('all')
params = ['d_market_belief', 
          'd_market_uncertainty', 
          'rejoice', 
          'regret', 
          'counterfactual_rejoice', 
          'counterfactual_regret']

ps = np.zeros((len(params), len(data)))
coeffs = np.zeros((len(params), len(data)))

for i, param in enumerate(params):
    for j, subject in enumerate(data):

        c_data = generate_subject_table(subject)
        bets = np.array(c_data.bet)
        d_bets = np.array(c_data.d_bet)


        for column in c_data.columns:
            if abs(c_data[column]).sum()>0:
                c_data[column] = zscore(c_data[column], nan_policy='omit')

        md = sm.OLS.from_formula('d_bet ~0 + ' + param, c_data, missing='drop')
        result = md.fit()

        indep_var = c_data[param]*c_data['prev_bet']*result.params[param] 
        df = pd.DataFrame({'resid': result.resid, 'indep_var': indep_var })

        new_md = sm.OLS.from_formula('resid ~ 0+ indep_var', df, missing='drop')
        new_result = new_md.fit()
        
        coeffs[i, j] = new_result.params[0]
        ps[i, j] = new_result.pvalues[0]

```

```python
df_coeffs = pd.DataFrame(coeffs.T, columns=params)
df_ps = pd.DataFrame(ps.T, columns=params)

plt.figure(dpi=400)
props = {'color': 'tab:blue'}
df_coeffs.boxplot(notch=True, showfliers=False, positions=np.arange(len(df_coeffs.columns))*2-0.4, widths=.6, medianprops=props, capprops=props, whis=[5,95])

props = {'color': 'tab:orange'}
df_ps.boxplot(notch=True, showfliers=False, positions=np.arange(len(df_ps.columns))*2+0.4, widths=.6, boxprops=props, medianprops=props, whiskerprops=props, capprops=props, whis=[5,95])

plt.grid(False)
plt.hlines(0, -0.5, len(df_ps.columns)*2, 'k', linestyles='dashed', linewidth=.7)
plt.xticks(rotation=45, ha="right")
plt.xticks(np.arange(len(df_ps.columns))*2, df_ps.columns)
plt.legend(handles=[mpatches.Patch(color='tab:blue', label='coefficient'), mpatches.Patch(color='tab:orange', label='p_value')])
plt.hlines(0.05, -0.5, len(df_ps.columns)*2, 'tab:orange', linewidth=.7)

plt.ylim(-2,2)
plt.show()
```
