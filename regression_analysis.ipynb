```python
%load_ext autoreload

%autoreload 2

import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore, sem
import pandas as pd
from tqdm import tqdm
import warnings
from statsmodels.tools.sm_exceptions import ConvergenceWarning
warnings.simplefilter('ignore', ConvergenceWarning)
warnings.simplefilter(action='ignore', category=FutureWarning)

# own modules
from utils import load_data_controls, load_data_patient

%run -i 'helperFiles/style_script.py'
```

```python
def convert_pvalue_to_asterisks(pvalue):
    
    if pvalue < 0.05:
        return '\u2217'
    return ''
```

```python
preds_controls, preds_patients = np.load('data/clusters_controls.npy'), np.load('data/clusters_patients.npy')
```

# Linear Regression

```python

all_data_controls = []
all_data_patients = []

for data, preds, all_dat in zip([load_data_controls('all'), load_data_patient('all')], [preds_controls, preds_patients], [all_data_controls, all_data_patients]):
    
    for cluster in [0,1]:
        
        cluster_data = pd.DataFrame()
        
        for i, subject in enumerate(data):
            
            if not cluster==preds[i]:
                continue
            subject['subj'] = i
            
            for column in subject.columns:
                if column in ['subj'] or abs(subject[column]).sum()==0:
                    continue
                subject[column] = zscore(subject[column], nan_policy='omit')
            cluster_data = pd.concat([cluster_data, subject])

        cluster_data.index = np.arange(len(cluster_data))
        all_dat.append(cluster_data)


all_data = [all_data_controls, all_data_patients]
groups = ['Controls', 'Patients']
labels = ['$b$', '$rej$', '$reg$', '$crej$', '$creg$', '$\mu_{GP}$']
```

```python
fig, ax = plt.subplots(1, 2, figsize=(figwidth,1.5), dpi=300, sharey=True, sharex=True)

for i, (data, group, color) in enumerate(zip(all_data, groups, colors)):
    
    coeffs = []
    errs = []
    p_vals = []
    
    for j, cluster_data in enumerate(data):
        
        md = sm.OLS.from_formula('d_bet ~ 0 + bet + rejoice + regret + counterfactual_rejoice + counterfactual_regret + d_market_belief ', 
                                 cluster_data, 
                                 missing='drop')
        result = md.fit()
        
        coeffs.append(result.params)
        errs.append(result.bse)
        p_vals.append(result.pvalues.values)
        
        print(group + ' - Cluster %d: BIC %.2f' % (j+1, result.bic)) 
        print(group + ' - Cluster %d: LL %d ' % (j+1, result.llf)) 
        print(group + ' - Cluster %d: R2 %.3f \n' % (j+1, result.rsquared_adj)) 

    pd.DataFrame({'C1': coeffs[0], 'C2': coeffs[1]}).plot(kind='bar', width=.5, ax=ax[i], color=color, yerr=errs).legend(loc='lower right', frameon=False, handlelength=1)
    ax[i].set_xticklabels(labels, rotation=0)
    ax[i].set_ylabel(r'$\beta$')
    
    for j, (p_val, coeff, err) in enumerate(zip(p_vals, coeffs, errs)):
        for p, c, e, x in zip(p_val, coeff, err, np.arange(len(labels))):
            y_pos = c + np.sign(c)*e
            if np.sign(c)>0:
                va = 'bottom'
            else:
                va='top'
            ax[i].text(x+(j*2-1)*0.125, y_pos, convert_pvalue_to_asterisks(p), ha='center', va=va, fontsize=7)

ax[0].text(0.5, -.5, '(a) Controls', ha='center', fontsize=12,transform=ax[0].transAxes)
ax[1].text(0.5, -.5 ,'(b) Patients', ha='center', fontsize=12,transform=ax[1].transAxes)
plt.ylim(-.6)
plt.savefig('figures/multiple_reg.pdf')
print('N parameters: ' + str(len(result.params)))
```

# Mixed effects model

```python
all_data_controls = []
all_data_patients = []

for data, preds, all_dat in zip([load_data_controls('all'), load_data_patient('all')], [preds_controls, preds_patients], [all_data_controls, all_data_patients]):
    
    for cluster in [0,1]:
        
        cluster_data = pd.DataFrame()
        
        for i, subject in enumerate(data):
            
            if not cluster==preds[i]:
                continue
            subject['subj'] = i
            cluster_data = pd.concat([cluster_data, subject])
            
        for column in cluster_data.columns:
            if column in ['subj', 'subj_n', 'diagnosis'] or abs(cluster_data[column]).sum()==0:
                continue
            cluster_data[column] = zscore(cluster_data[column], nan_policy='omit')
        
        cluster_data.index = np.arange(len(cluster_data))
        all_dat.append(cluster_data)
    
all_data = [all_data_controls, all_data_patients]
groups = ['Controls', 'Patients']
labels = ['$b$', '$rej$', '$reg$', '$crej$', '$creg$', '$\mu_{GP}$']
```

## only random intercepts

```python
fig, ax = plt.subplots(1, 2, figsize=(figwidth,1.5), dpi=300, sharey=True, sharex=True)

for i, (data, group, color) in enumerate(zip(all_data, groups, colors)):
    
    coeffs = []
    errs = []
    p_vals = []
    
    for j, cluster_data in enumerate(data):
        
        md = smf.mixedlm('d_bet ~ 0 + bet + rejoice + regret + counterfactual_rejoice + counterfactual_regret + d_market_belief', 
                         cluster_data, 
                         missing = "drop", 
                         groups=cluster_data["subj"])
        result = md.fit(reml = False, method = 'BFGS')
        
        coeffs.append(result.params[['bet', 'rejoice', 'regret', 'counterfactual_rejoice', 'counterfactual_regret', 'd_market_belief']])
        errs.append(result.bse[['bet', 'rejoice', 'regret', 'counterfactual_rejoice', 'counterfactual_regret', 'd_market_belief']])
        p_vals.append(result.pvalues[['bet', 'rejoice', 'regret', 'counterfactual_rejoice', 'counterfactual_regret', 'd_market_belief']].values)
        
        print(group + ' - Cluster %d: BIC %.2f' % (j+1, result.bic)) 
        print(group + ' - Cluster %d: LL %d' % (j+1, result.llf)) 

    pd.DataFrame({'C1': coeffs[0], 'C2': coeffs[1]}).plot(kind='bar', width=.5, ax=ax[i], color=color, yerr=errs).legend(loc='lower right', frameon=False, handlelength=1)
    ax[i].set_xticklabels(labels, rotation=0)
    ax[i].set_ylabel(r'$\beta$')
    
    for j, (p_val, coeff, err) in enumerate(zip(p_vals, coeffs, errs)):
        for p, c, e, x in zip(p_val, coeff, err, np.arange(len(labels))):
            y_pos = c + np.sign(c)*e
            if np.sign(c)>0:
                va = 'bottom'
            else:
                va='top'
            ax[i].text(x+(j*2-1)*0.125, y_pos, convert_pvalue_to_asterisks(p), ha='center', va=va, fontsize=7)

ax[0].text(0.5, -.5, '(a) Controls', ha='center', fontsize=12,transform=ax[0].transAxes)
ax[1].text(0.5, -.5 ,'(b) Patients', ha='center', fontsize=12,transform=ax[1].transAxes)
plt.ylim(-.7)
plt.savefig('figures/me_intercept.pdf')
plt.savefig('figures/me_intercept.png')
print('N parameters: ' + str(len(result.params)))
```
